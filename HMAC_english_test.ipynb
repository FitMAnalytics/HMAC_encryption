```python
from utils import (
    tokenize_address, 
    get_hmac_1grams, get_hmac_2grams, 
    jaro_similarity_hmac, jaro_winkler_hmac, 
    levenshtein_hmac, 
    get_qgrams_from_hash_list, calculate_qgram_similarity, 
    calculate_jaccard)
```

```python
address_1 = "712 5th Ave, New York, NY 10019"
address_2 = "721 25th Ave, New York, NY 10027"

hash_1 = tokenize_address(address_1,language="en")
hash_2 = tokenize_address(address_2,language="en")

hmac_1_1grams = get_hmac_1grams(hash_1)
hmac_2_1grams = get_hmac_1grams(hash_2)
```

```python
print("-------HMAC 2-grams Encoding---------")
hmac_1 = get_hmac_2grams(hash_1)
hmac_2 = get_hmac_2grams(hash_2)

print(f"Jaro Similarity: {jaro_similarity_hmac(hmac_1, hmac_2):.4f}")
print(f"Jaro Winkler Similarity: {jaro_winkler_hmac(hmac_1, hmac_2):.4f}")

levenshtein_similarity = 1-levenshtein_hmac(hmac_1, hmac_2)/max(len(hmac_1), len(hmac_2))

print(f"Levenshtein Similarity: {levenshtein_similarity:.4f}")

q = 2
hmac_1_qgrams = get_qgrams_from_hash_list(hmac_1, q=q)
hmac_2_qgrams = get_qgrams_from_hash_list(hmac_2, q=q)

qgram_similarity = calculate_qgram_similarity(hmac_1_qgrams, hmac_2_qgrams)
print(f"Qgram (with q = 2) Similarity: {qgram_similarity:.4f}")

q_gram_jaccard_similarity = calculate_jaccard(set(hmac_1_qgrams), set(hmac_2_qgrams))
print(f"Q-gram Jaccard Similarity: {q_gram_jaccard_similarity:.4f}")

print(f"Jaccard Similarity: {calculate_jaccard(set(hmac_1), set(hmac_2)):.4f}")
```
