```python
from utils import (
    tokenize_English_address,
    get_hmac_1grams, get_hmac_2grams,
    get_hmac_1grams_en_str,
    tokenize_hmac,
    jaro_similarity_hmac, jaro_winkler_hmac, 
    levenshtein_hmac, 
    get_qgrams_from_hash_list, calculate_qgram_similarity, 
    calculate_jaccard)
```

```python
address_1 = "712 5th Ave, New York, NY 10019"
address_2 = "721 25th Ave, New York, NY 10027"

tokens_1 = tokenize_English_address(address_1)
tokens_2 = tokenize_English_address(address_2)

print(f"address_1: {address_1} \n\naddress_1_tokenized: {tokens_1} \n\naddress_2: {address_2}\n\naddress_2_tokenized: {tokens_2}\n\n")

truncate_length = 12
address_1_hmac_1grams = get_hmac_1grams_en_str(tokens_1, truncate_switch=True, truncate_length=truncate_length)
address_2_hmac_1grams = get_hmac_1grams_en_str(tokens_2, truncate_switch=True, truncate_length=truncate_length)

print(f"address_1_hmac_1grams: {address_1_hmac_1grams} \n\naddress_2_hmac_1grams: {address_2_hmac_1grams}\n\n")

print(f"address_1 tokenized hmac: {tokenize_hmac(address_1_hmac_1grams, truncate_length)} \n\naddress_2 tokenized hmac: {tokenize_hmac(address_2_hmac_1grams, truncate_length)}\n\n")

```
